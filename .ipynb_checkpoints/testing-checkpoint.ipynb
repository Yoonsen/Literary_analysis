```python
def related(idd):
    return requests.get('https://api.nb.no:443/catalog/v1/metadata/{idd}/struct'.format(idd=idd)).text
```

```python
xmls = related('URN:NBN:no-nb_digibok_2011110906074')
```

```python
isinstance(xmls,str)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(xmls)
soup.prettify()
# '<html>\n <head>\n </head>\n <body>\n  <a href="http://example.com/">\n...'

print(soup.prettify())
```
